## Section 3: SSH Key Setup (10 min)

### Why SSH Keys?
SSH keys provide secure, passwordless authentication to HPC systems.

### Step 1: Generate SSH Key (if you don't have one)
Open a terminal on your local machine:

```bash
# Generate ED25519 key (recommended)
ssh-keygen -t ed25519 -C "your_email@hi.is" -f ~/.ssh/jureca_key

# Alternative: Generate RSA key (4096 bits)
ssh-keygen -t rsa -b 4096 -C "your_email@hi.is" -f ~/.ssh/jureca_key
```

**Press Enter** when asked for passphrase (or set one for extra security)

### Step 2: Copy Public Key
```bash
# Display your public key
cat ~/.ssh/jureca_key.pub
```
Copy the entire output (starts with `ssh-ed25519` or `ssh-rsa`)

### Step 3: Upload to Judoor
1. Go to: https://judoor.fz-juelich.de/profile
2. Navigate to **"SSH Public Keys"** section
3. Click **"Add SSH Key"**
4. Paste your public key
5. Give it a recognizable name (e.g., "My Laptop")
6. Click **"Save"**

### Step 4: Configure SSH Client
Create/edit `~/.ssh/config` on your local machine:

```bash
Host jureca
    HostName jureca-login.fz-juelich.de
    User YOUR_JUDOOR_USERNAME
    IdentityFile ~/.ssh/jureca_key
    ServerAliveInterval 60
```

Replace `YOUR_JUDOOR_USERNAME` with your Judoor username.

## ðŸ”µ Advanced Topic 1: SSH Config File

### Why Use SSH Config?
Instead of typing long SSH commands, create shortcuts!

### Create/Edit `~/.ssh/config`
```bash
# JURECA Login
Host jureca
    HostName jureca-login.fz-juelich.de
    User YOUR_USERNAME
    IdentityFile ~/.ssh/jureca_key
    ServerAliveInterval 60
    ServerAliveCountMax 10

# JURECA with Port Forwarding (for Jupyter)
Host jureca-jupyter
    HostName jureca-login.fz-juelich.de
    User YOUR_USERNAME
    IdentityFile ~/.ssh/jureca_key
    LocalForward 8888 localhost:8888
```

### Usage
```bash
# Instead of: ssh -i ~/.ssh/jureca_key username@jureca-login.fz-juelich.de
# Simply type:
ssh jureca
```

## ðŸ”µ Advanced Topic 2: Terminal Multiplexers (tmux/screen)

### Why Use tmux?
- Keep sessions running even if you disconnect
- Multiple terminals in one SSH session
- Essential for long-running processes

### Basic tmux Commands
```bash
# Start new session
tmux new -s ml_session

# Detach (keep running): Ctrl+b then d

# List sessions
tmux ls

# Reattach to session
tmux attach -t ml_session

# Kill session
tmux kill-session -t ml_session
```

### Quick tmux Shortcuts
- `Ctrl+b c` - Create new window
- `Ctrl+b n` - Next window
- `Ctrl+b p` - Previous window
- `Ctrl+b %` - Split vertically
- `Ctrl+b "` - Split horizontally

## ðŸ”µ Advanced Topic 3: SLURM Job Submission

### Interactive Job
```bash
# Request interactive session
srun --partition=dc-cpu --nodes=1 --ntasks=1 --time=01:00:00 --pty /bin/bash

# Request GPU
srun --partition=dc-gpu --nodes=1 --gres=gpu:1 --time=01:00:00 --pty /bin/bash
```

### Batch Job
Create a file `test_job.sh`:
```bash
#!/bin/bash
#SBATCH --partition=dc-cpu
#SBATCH --nodes=1
#SBATCH --ntasks=1
#SBATCH --time=00:10:00
#SBATCH --job-name=test_job
#SBATCH --output=test_%j.out
#SBATCH --error=test_%j.err

echo "Job started at $(date)"
echo "Running on node: $(hostname)"
sleep 30
echo "Job completed at $(date)"
```

Submit with:
```bash
sbatch test_job.sh
```

### Monitor Jobs
```bash
# Check your jobs
squeue -u $USER

# Check all jobs in partition
squeue -p dc-cpu

# Cancel a job
scancel JOB_ID

# Get job info
scontrol show job JOB_ID
```

## ðŸ”µ Advanced Topic 4: File Transfer Methods

### 1. SCP (Secure Copy)
```bash
# Copy from local to JURECA
scp myfile.txt jureca:$PROJECT_training2600/data/

# Copy from JURECA to local
scp jureca:$PROJECT_training2600/results/output.txt ./

# Copy directory recursively
scp -r mydir/ jureca:$PROJECT_training2600/
```

### 2. Rsync (Recommended for Large Files)
```bash
# Sync directory (only copies changes)
rsync -avzP mydir/ jureca:$PROJECT_training2600/mydir/

# Options:
# -a: archive mode (preserves permissions, timestamps)
# -v: verbose
# -z: compress during transfer
# -P: show progress and resume partial transfers
```

### 3. Globus (For Very Large Datasets)
- Web-based transfer tool
- Handles TB-scale data
- Automatic retry on failures
- Visit: https://www.globus.org/
